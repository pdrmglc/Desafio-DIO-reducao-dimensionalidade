```python
from PIL import Image

# Abrir a imagem
imagem_colorida = Image.open("imagem_desafio_cinza_binario.png")
from PIL import Image

# Converter para tons de cinza
if imagem_colorida.mode != "RGB":
    imagem_colorida = imagem_colorida.convert("RGB")

# Obter dimenses da imagem
largura, altura = imagem_colorida.size

# Criar uma nova imagem em tons de cinza
imagem_cinza = Image.new("L", (largura, altura))

# Percorrer cada pixel e converter para tons de cinza
for y in range(altura):
    for x in range(largura):
        # Obter o valor RGB do pixel
        r, g, b = imagem_colorida.getpixel((x, y))
        
        # Calcular a mdia ponderada dos canais RGB
        tom_de_cinza = int(0.2989 * r + 0.587 * g + 0.114 * b)
        
        # Definir o novo valor do pixel na imagem em tons de cinza
        imagem_cinza.putpixel((x, y), tom_de_cinza)

# Salvar a imagem em tons de cinza
imagem_cinza.save("imagem_cinza.png")

```

```python
from PIL import Image

# Abrir a imagem em tons de cinza
imagem_cinza = Image.open("imagem_cinza.png")

# Obter dimenses da imagem
largura, altura = imagem_cinza.size

# Definir um limiar (threshold)
limiar = 128

# Criar uma nova imagem em binrio preto e branco
imagem_binaria = Image.new("1", (largura, altura))

# Percorrer cada pixel e converter para binrio preto e branco
for y in range(altura):
    for x in range(largura):
        # Obter o valor do pixel em tons de cinza
        tom_de_cinza = imagem_cinza.getpixel((x, y))
        
        # Verificar se o valor do pixel  maior que o limiar
        if tom_de_cinza > limiar:
            # Definir o pixel como branco (255)
            imagem_binaria.putpixel((x, y), 255)
        else:
            # Definir o pixel como preto (0)
            imagem_binaria.putpixel((x, y), 0)

# Salvar a imagem em binrio preto e branco
imagem_binaria.save("imagem_binaria.png")
```

```python
# Abrir a imagem em binrio preto e branco
imagem_binaria = Image.open("imagem_binaria.png")

# Exibir a imagem
plt.imshow(imagem_binaria, cmap="gray")
plt.axis('off')
plt.show()
```
